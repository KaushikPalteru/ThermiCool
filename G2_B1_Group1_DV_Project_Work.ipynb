# 3.DURING WHICH TIME OF THE YEAR,GAMES ZONE HAS THE HIGHEST NUMBER OF CUSTOMERS?

```python
c=[]
m=[31,28,31,30,31,30,31,31,30,31,30,31]
games=df['games'].tolist()
k=0
for i in m:
  sum=0
  for j in range(0,i):
    sum=sum+games[k]
    k=k+1
  c.append(sum)
print(c)
```

```python
labels=['jan','feb','mar','apr','may','jun','jul','aug','sep','out','nov','dec']
plt.pie(c,labels=labels, startangle=60, autopct="%0.1f%%", wedgeprops={'edgecolor':'black'})
plt.show()
```

# 5.WHICH WEEK OF THE MONTH JANUARY HAS THE HIGHEST SALE?

```python
c=[]
T=[]
d=[7,7,7,7]
sal=df['sales'].tolist()
date=df['DATE'].tolist()
k=0
for i in d:
  f=[]
  g=[]
  for j in range(0,i):
      f.append(sal[k])
      g.append(date[k])
      k=k+1 
  c.append(f)
  T.append(g)     
print(T)
print(c)
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
fig4, ax4 = plt.subplots()

ax1.bar(T[0],c[0],color='#800000' )
ax2.bar(T[1],c[1],color='#CC7722' )
ax3.bar(T[2],c[2],color='#808000')
ax4.bar(T[3],c[3],color='#A95C68')

ax1.set_title("sales in first week")
ax1.set_ylabel("no. of items sold")
ax1.set_xlabel("date")


ax2.set_title("sales in second week")
ax2.set_ylabel("no. of items sold")
ax2.set_xlabel("date")

ax3.set_title("sales in third week")
ax3.set_ylabel("no. of items sold")
ax3.set_xlabel("date")


ax4.set_title("sales in fourth week")
ax4.set_ylabel("no. of items sold")
ax4.set_xlabel("date")

plt.tight_layout()
plt.show()
```

#6. ON WHICH DAY OF THE YEAR DOES THE GIFTSTORE HAVE THE HIGHEST SALE?

```python
c=[]
d=[]
m=[31,28,31,30,31,30,31,31,30,31,30,31]
giftstore=df['gift store'].tolist()
date=df['DATE'].tolist()
k=0
for i in m:
  for j in range(0,i):
    if giftstore[k]>=900:
      c.append(giftstore[k])
      d.append(date[k])
    k=k+1  
print(d)
print(c)
```

```python
plt.figure(figsize=(10,8))
plt.stackplot(d,c,color='blue')
plt.xlabel("date")
plt.ylabel("num of items sold")
plt.title("high sales in giftstore on these particular days")
plt.show()
```
